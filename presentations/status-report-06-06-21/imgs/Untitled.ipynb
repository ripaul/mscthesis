```python
import numpy as np
from math import factorial
import matplotlib as mpl
import matplotlib.pyplot as plt

dpi = 300
```

```python
def hanging_line(point1, point2, h=1):
    m = 0.5*(point1[0] + point2[0])
    a = -h / (point1[0] - m)**2
    x = np.linspace(point1[0], point2[0], 100)
    y = a*(x - m)**2 + h

    return (x,y)

def choose(n, k):
    return factorial(n) / factorial(k) / factorial(n - k)
```

```python
X = np.array(range(0,4)) * 50

fig, axs = plt.subplots(2, 1, dpi=dpi)
for ax in axs:
    ax.set_xlim([np.min(X)-20, np.max(X) + 20])

for i, x in enumerate(X):
    circle = plt.Circle((x, 0), 10, edgecolor='black', facecolor='white')
    enum = str(i)#-int(len(X)/2))
    axs[1].text(x, 0, enum, ha='center', va='center')
    axs[1].add_patch(circle)

style = "Simple, tail_width=0.5, head_width=4, head_length=8"
kw = dict(arrowstyle=style, color="k")

for i, x in enumerate(X[:-1]):
    patch = mpl.patches.FancyArrowPatch(
        (X[i]+9*np.cos(-.2*np.pi), 9*np.sin(-.2*np.pi)), 
        (X[i+1]+9*np.cos(-.9*np.pi), 9*np.sin(-.8*np.pi)), 
        connectionstyle="arc3,rad=.3", **kw)
    axs[1].add_patch(patch)
    patch = mpl.patches.FancyArrowPatch(
        (X[i+1]+9*np.cos(.8*np.pi), 9*np.sin(.8*np.pi)), 
        (X[i]+9*np.cos(.2*np.pi), 9*np.sin(.2*np.pi)), 
        connectionstyle="arc3,rad=.3", **kw)
    axs[1].add_patch(patch)
    axs[1].text(0.5*(X[i]+X[i+1]), 17, r"$p_{"+str(i+1)+str(i)+"}$", ha='center', va='center')
    axs[1].text(0.5*(X[i]+X[i+1]), -17, r"$p_{"+str(i)+str(i+1)+"}$", ha='center', va='center')
    #x_line, y_line = hanging_line((X[i], 0), (X[i+1], 0), h=10)
    #axs[1].plot(x_line, y_line, zorder=-10, c="black")
    #axs[1].plot(x_line, -y_line, zorder=-10, c="black")

axs[1].set_ylim([-20, 20])
axs[1].set_aspect(1)
axs[1].axis("off")

axs[0].spines['right'].set_visible(False)
axs[0].spines['bottom'].set_visible(False)
axs[0].spines['top'].set_visible(False)
axs[0].set_xticks([])
axs[0].set_xticklabels([])

p = 0.5
dist = []

for i, x in enumerate(X):
    dist.append(choose(len(X)-1, i) * p**i * ((1-p)**(len(X)-1-i)))

offset = 20 * np.ones(X.shape)

axs[0].bar(X, dist, width=20, label=r"$\pi(x)$")
axs[0].legend(loc='upper right')

fig.subplots_adjust(left=0,
                    bottom=0, 
                    right=1, 
                    top=1, 
                    wspace=0, 
                    hspace=-0.35)
#fig.tight_layout()
fig.savefig("binmoialchaindistribution.png", bbox_inches='tight')
plt.show()
```

```python
def gaussian(x, mu, sigma, normalize=True):
    Z = (sigma*np.sqrt(2*np.pi)) if normalize else 1
    return np.exp(-0.5*(x-mu)**2/sigma**2) / Z

x = np.linspace(-4, 6, 200)
y = gaussian(x, 0, 1)

p1 = gaussian(x, -1.5, 4, False)
p2 = gaussian(x, -1.5, 1, False)
p3 = gaussian(x, -1.5, 0.2, False)

fig, ax = plt.subplots(1, 1, dpi=dpi)

ax.plot(x, y, label=r'target $\mathcal{N}(0, 1)$')
ax.plot([-1.5, -1.5], [0, 0.4], color='#888', linestyle='dashed', label=r'current state $\theta$')
ax.fill_between(x, np.zeros(x.shape), 0.1*p1, alpha = 0.5, color='C1', label=r'proposal $\propto \mathcal{N}(\theta, s=4)$')
ax.fill_between(x, np.zeros(x.shape), 0.25*p2, alpha = 0.5, color='C2', label=r'proposal $\propto \mathcal{N}(\theta, s=1)$')
ax.fill_between(x, np.zeros(x.shape), 0.4*p3, alpha = 0.5, color='C3', label=r'proposal $\propto \mathcal{N}(\theta, s=0.2)$')

ticks = [-4, -2, -1.5, 0, 2, 4, 6]
ax.set_xticks(ticks)
ticklabels = [-4, -2, r'$\theta$', 0, 2, 4, 6]
ax.set_xticklabels(ticklabels)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.text(2, 0.21, "proposals are not correclty normalized!", fontsize=7)

ax.set_xlabel(r"$\Theta$")
ax.set_ylabel(r"$f(\Theta)$")

ax.legend()
fig.savefig("stepsizes.png")
```
